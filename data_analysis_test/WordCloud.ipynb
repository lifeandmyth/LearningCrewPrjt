```python
import numpy as np
from PIL import Image
icon = Image.open('../data_analysis_test/images/iphone-160307_1920.png').convert("RGBA")
# icon = Image.open('images/heart.png').convert("RGBA")
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon,icon)
mask = np.array(mask)
```

```python
from collections import Counter
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
dic1 = dict(zip(df['words'], df['ranks']))
# dic1
```

```python
f_path = '../data_analysis_test/fonts/NanumMyeongjoExtraBold.ttf'

wordcloud = WordCloud(font_path=f_path,
                      background_color='white', 
                      width=800, 
                      height=800,
                      mask=mask, colormap='rainbow').generate_from_frequencies(dic1)
fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='nearest')
plt.axis('off')
plt.savefig(f'../data_analysis_test/images/created/wordcloud_cellphone.png', dpi=300)
```
